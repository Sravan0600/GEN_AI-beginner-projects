# Extract Text from the PDF

```python
from PyPDF2 import PdfReader

# Step 1: Extract text from the uploaded PDF
def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

# Replace this with your actual file name after upload
pdf_filename = "grokking-algorithms-illustrated-programmers-curious.pdf"

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_filename)
```

```python
# Print the first 1000 characters of the extracted text
print(pdf_text[:1000])  # Check the beginning of the extracted text
```

# Split the Extracted Text into Chunks

```python
import re

# Split by sentences using regular expression
text_chunks = re.split(r'(?<=[.!?]) +', pdf_text)  # Split on sentence endings followed by a space
text_chunks = [chunk.strip() for chunk in text_chunks if chunk]  # Clean up any empty chunks

# Limit chunk size if needed
max_chunk_size = 100 # Add this line to define the variable max_chunk_size and set it to 100
text_chunks = [' '.join(chunk.split()[:max_chunk_size]) for chunk in text_chunks]

# Check the number of chunks and display a few sample chunks
print(f"Total chunks: {len(text_chunks)}")
print(f"First chunk: {text_chunks[0]}")
```

# Convert the Text Chunks into Embeddings

```python
from langchain.embeddings import SentenceTransformerEmbeddings
from langchain.docstore.document import Document
from langchain.vectorstores import FAISS

# Step 2: Convert text chunks to Document objects and generate embeddings
documents = [Document(page_content=chunk) for chunk in text_chunks]

# Load the embedding model
embeddings = SentenceTransformerEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')

# Store the embeddings in FAISS
vector_store = FAISS.from_documents(documents, embeddings)
```

# Query the Vector Store

```python
# Step 3: Function to query FAISS vector store and return relevant chunks
def query_pdf(user_query, vector_store, k=5):
    # Perform similarity search in FAISS
    results = vector_store.similarity_search(user_query, k=k)  # k is how many results to return
    return results

# User query example
user_query = "Explain the basic concepts of algorithms"

# Perform the query
search_results = query_pdf(user_query, vector_store, k=5)  # Get top 5 relevant chunks

# Step 4: Display the search results
for i, result in enumerate(search_results):
    print(f"Result {i+1}:\n{result.page_content}\n")
```
